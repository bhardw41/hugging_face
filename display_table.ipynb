```python
from pathlib import Path

folder_path = Path("EXAMPLES_Output")
file_pattern = "slurm-*_*.out"

fitness_lines = []
file_list =[]
for file_path in folder_path.glob(file_pattern):
    temp_list =[]
    file_list.append(file_path)
    with open(file_path, "r") as file:
        for line in file:
            if "#FITNESS" in line:
                temp_list.append(line.strip("#FITNESS"))
    fitness_lines.append(temp_list)
file_list
```

```python
fitness_values = []
count = 0
for each_image in fitness_lines:
    print(file_list[count])
    each_fitness = []
    for each_model in each_image:
        if "ERROR" in each_model:
            each_fitness.append('err')
        elif each_model[1] == '[':
            if float(each_model[2:5]) == 0.0:
                each_fitness.append(float(each_model[2:9]))
            elif float(each_model[2:5]) == 1.5:
                each_fitness.append('INF')
            else:
                each_fitness.append(float(each_model[2:5]))
    fitness_values.append(each_fitness)
    count +=1
            
       
            
        
```

```python
for each in fitness_values:
    ind = fitness_values.index(each)
    if each == []:
        fitness_values.remove(each)
        file_list.pop(ind)
 
```

```python
import pandas as pd
df = pd.DataFrame(fitness_values, index = file_list)
df.index.names = ["Images"]
pd.options.display.float_format = '{:.3f}'.format

```

```python
pd.set_option('display.max_columns', 230) #replace n with the number of columns you want to see completely
pd.set_option('display.max_rows', 230) 


```

```python
selected_columns = []
for column in df:
    # Count the occurrences of 'Navya' in the column
    err_count = df[column].eq('err').sum()

    point_count = df[column].eq('INF').sum()
    if err_count < len(df) / 2 and point_count < 5:
        selected_columns.append(column)

# Print the DataFrame with the selected columns
new_df = df[selected_columns]
```

```python
def bold_value(value):
#     threshold = 20  # Set your desired threshold here
    try:
        if value < 1:
            return 'font-weight: bold; {:.3f}'.format(value)
        else:
            return '{:.3f}'.format(value)
    except:
        return
```

```python
new_df = new_df.style.applymap(bold_value)
pd.options.display.float_format = '{:.3f}'.format
new_df
```
